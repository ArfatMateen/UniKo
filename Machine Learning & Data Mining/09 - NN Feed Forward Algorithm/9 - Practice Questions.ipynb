## NN Forward Pass

![title](Neural_Network_Graph.png)

```python
inputs = np.array([0.3, 0.4])
input_weights = np.array([0.5, 0.1])
layers_weights = np.array([0.85, 0.45])
bias = np.array([0.33, 0.1])
```

```python
x = (inputs[0] * input_weights[0]) + (inputs[1] * input_weights[1]) + bias[0]
x
```

```python
out_1 = 1 / (1 + np.exp(-x))
out_1
```

```python
in_2 = (out_1 * layers_weights[0]) + bias[1]
in_2
```

```python
out_2 = 1 / (1 + np.exp(-in_2))
out_2
```

```python
in_output = out_2 * layers_weights[1]
in_output
```

```python
output = 1 / (1 + np.exp(-in_output))
output
```

```python
target = 0.01
```

```python
0.5 * np.square(target - output)
```

## Map higher dimension

```python
X = np.array([[0,6], [1,5], [3,3], [4,2], [6,0]])
```

```python
X
```

```python
Y = np.array([0, 1, 1, 1, 0])
```

```python
Y
```

```python
plt.scatter(X[:, 0], X[:, 1], c=Y)

plt.xlim(0, 7)
plt.ylim(0, 7)
```

```python
def phi(X):
    """Non-linear Transformation"""
    
    X1 = X[:, 0] ** 2
    X2 = X[:, 1] ** 2
    X3 = np.sqrt(2) * X[:, 0] * X[:, 1]
    
    X_ = np.zeros((X.shape[0], 3))

#     X_[:, :-1] = X
    X_[:, -3] = X1
    X_[:, -2] = X2
    X_[:, -1] = X3
    
    return X_
```

```python
X_ = phi(X)
```

```python
X_.shape
```

```python
X_
```

```python
def plot3d(X):
    fig = plt.figure(figsize=(12, 12))
    ax = fig.add_subplot(111, projection='3d')
    
    X1 = X[:, 0]
    X2 = X[:, 1]
    X3 = X[:, 2]
    
    ax.scatter(X1, X2, X3, zdir='z', c = Y, s = 20, depthshade=False)
    
    ax.set_xlabel('$X$', fontsize=30)
    ax.set_ylabel('$Y$', fontsize=30)
    ax.set_zlabel('$Z$', fontsize=30)
    
    plt.show()
```

```python
plot3d(X_)
```
