```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1.1 SIN / COSIN Plot

```python
rand_nums = np.random.randint(0,80,10)
```

```python
#convert degrees to radians
X = [(float(x) * np.pi) / 180 for x in rand_nums]
```

```python
SIN = np.sin(X)
```

```python
COSIN = np.cos(X)
```

```python
fig = plt.figure(figsize=(10, 6))

ax = fig.add_axes([0,0,1,1])

ax.scatter(rand_nums, SIN, label="SIN", color='tab:blue')
ax.scatter(rand_nums, COSIN, label="COSIN", color='tab:orange')

ax.set(xlabel="X values", ylabel="SIN / COSIN values", title="SIN / COSIN Plot")
ax.grid(alpha = 0.2)
ax.legend()
```

```python
fig.savefig("SinCos Plot.png", 
            dpi=100, 
            bbox_inches ="tight",
            pad_inches = 0.5,
            facecolor ="w")
```

## 1.2 Frequency Distribution

```python
file = open("sample.txt")

text = file.read()
file.close()
```

```python
char_count = {}
for c in set(text):
    char_count[c] = text.count(c)
```

```python
sorted_char_count = dict(sorted(char_count.items()))
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_axes([0,0,1,1])

ax.bar(sorted_char_count.keys(), sorted_char_count.values())

ax.set(xlabel="Character Distribution", ylabel="Frequency", title="Frequency Distribution")
ax.grid(alpha = 0.2)
```

```python
fig.savefig("Frequency Distribution Plot.png", 
            dpi=100, 
            bbox_inches ="tight",
            pad_inches = 0.5,
            facecolor ="w")
```
