```python
import string
import re
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gen_word(length , chars , p_chars):
    word = ''.join([ chars[r] for r in np.random.choice(len(p_chars), length , p=p_chars) ])
    return word
```

```python
def main(input_file):
    data = open(input_file, 'r').read().lower()
    text = re.sub(r'[\.]+', '', data)
    
    # generate characters, calculate probability of each character
    chars = string.ascii_lowercase + string.digits
    freq_chars = np.array([ text.count(c) for c in chars ])
    p_chars = freq_chars / sum(freq_chars)

    # generate word lengths, calculate probability of word length
    word_lengths = [ len(w) for w in text.split() ]
    uniq_word_lengths = np.unique(word_lengths)
    num_words = len(word_lengths)
    p_word_lengths = [ word_lengths.count(l) / num_words for l in uniq_word_lengths ]
    
    # generate text
    word_lengths_gen = np.random.choice(uniq_word_lengths, 5000, p=p_word_lengths)
    text_gen = ' '.join([ gen_word(l, chars, p_chars) for l in word_lengths_gen ])

    # print generated text
    print (text_gen)

    
    # calculating probability of generated words' length for generating the plot
    word_lengths_gen = [ len(w) for w in text_gen.split() ]
    p_word_lengths_gen = [ word_lengths_gen.count(l) / 5000 for l in uniq_word_lengths ]

    # plotting
    fig, ax = plt.subplots()
    bar1 = ax.bar(uniq_word_lengths-0.15, p_word_lengths, width=0.3, color='b',align='center')
    bar2 = ax.bar(uniq_word_lengths+0.15, p_word_lengths_gen, width=0.3, color='r',align='center')

    plt.xticks(np.arange(max(uniq_word_lengths)) + 1)
    ax.set_xlabel('Word Length')
    ax.set_ylabel('Probability')
    ax.set_title('Distribution of Word Lengths')

    ax.legend((bar1[0], bar2[0]), ('Original', 'Generated'))
    plt.savefig(output_plot, format='png')

if __name__ == '__main__':
    input_file = 'Text_sample.txt'
    output_plot = 'Figure_1.png'
```

```python
main(input_file)
```
